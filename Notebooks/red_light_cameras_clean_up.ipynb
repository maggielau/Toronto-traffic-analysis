```python
import pandas as pd
import numpy as np
import datetime

#file_path = '/content/drive/MyDrive/Data Files/Data Files/'  # Path specific to Dennis
file_in = 'Red Light Cameras Data.csv'
```

```python
file_content = pd.read_csv(file_in,
                           nrows = None,
                           index_col='ID',
                           )
file_content
```

```python
file_content[(file_content['_id']==20823)|(file_content['_id']==20891)][['NAME','MAIN']]
```

```python
file_content.info()
```

```python
new = file_content['geometry'].str.split(":",expand=True)
new2=new.iloc[:,2].str.split(",",expand=True)
```

```python
def longitude(x):
    return x[0][2:]

def latitude(x):
    return x[1][:-2]

def activation_day(x):
    return pd.to_datetime(x['ACTIVATION_DATE'][:10] + ' '+ x['ACTIVATION_DATE'][12:])
```

```python
lon=new2.apply(longitude,axis=1)
lon.astype(float)
```

```python
lat=new2.apply(latitude,axis=1)
lat.astype(float)
```

```python
file_content['LONGITUDE']=lon
file_content['LATITUDE']=lat
```

```python
file_content["ACTIVATION_TIME"] = file_content.apply(activation_day,axis=1)
```

```python
file_content.rename(columns={"_id": "CAMERA_ID"},inplace=True)
```

```python
red_light=file_content[['CAMERA_ID','INTERSECTION_ID','LINEAR_NAME_FULL_1','LINEAR_NAME_FULL_2','LONGITUDE','LATITUDE','ACTIVATION_TIME']]
red_light
```
