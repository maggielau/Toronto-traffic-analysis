```python
# Initialize the environment

import pandas as pd
import numpy as np
import datetime
from IPython.display import display
import plotly.express as px
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh') 
import seaborn as sns

file_path = './Data Files/'
case_file = 'NCDB_Cases.csv'
vehicle_file = 'NCDB_Vehicles.csv'
people_file = 'NCDB_People.csv'
```

```python
# Read the case data into a dataframe

case_df = pd.read_csv(file_path + case_file, 
                      dtype={'C_MNTH':str,
                             'pd_WDAY':str})
case_df.head(5)
```

```python
# Read the people data into a dataframe

people_df = pd.read_csv(file_path + case_file,
                        dtype={'C_MNTH':str,
                               'pd_WDAY':str})
people_df.head(5)
```

```python
case_df.groupby(['INVOLVED']).aggregate({'C_CASE': 'count'}).\
    plot(kind='bar', width=0.75, rot=90, figsize=(15, 10)); 
```

```python
case_df.groupby(['C_HOUR']).\
        aggregate({'C_CASE': 'count'}).\
        plot(kind='bar', 
             width=0.75, 
             rot=90, 
             figsize=(15, 10)
            ); 
```
